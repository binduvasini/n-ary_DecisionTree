```python
import pandas as pd
from decision_tree import decision_tree

zoo = pd.read_csv('zoo.csv')
X = zoo.iloc[:,1:-1]
Y = zoo.iloc[:,-1]
XList = X.values.tolist()
YList = Y.values.tolist()
```

```python
zoo.head()
```

```python
#building the tree with the criterion passed as a parameter
dd = decision_tree('gini')
tree = dd.fit(XList,YList)
tree
```

```python
#Predicting using the same trained dataset
hypo = dd.predict(XList)
hypo
```
